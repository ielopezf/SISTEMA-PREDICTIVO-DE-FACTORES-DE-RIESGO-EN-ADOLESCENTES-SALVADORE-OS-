# 8. Análisis Demográfico

```python
# ==========================================================
# FUNCIONES DE VISUALIZACIÓN
# ==========================================================

def grafico_barras(df, columna, titulo, xlabel="", ylabel="Cantidad"):
    """
    Genera un gráfico de barras para variables categóricas.
    """

    datos = df[columna].value_counts(dropna=False).sort_index()

    plt.figure(figsize=(8,5))

    plt.bar(
        datos.index.astype(str),
        datos.values
    )

    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.grid(axis="y", alpha=0.3)

    plt.tight_layout()

    plt.show()
```

```python
def histograma(df, columna, titulo):

    plt.figure(figsize=(9,5))

    plt.hist(
        df[columna].dropna(),
        bins=20,
        edgecolor="black"
    )

    plt.title(titulo, fontsize=14)

    plt.xlabel(columna)

    plt.ylabel("Frecuencia")

    plt.grid(alpha=0.25)

    plt.tight_layout()

    plt.show()
```

```python
def boxplot(df, columna, titulo):

    plt.figure(figsize=(9,2.8))

    plt.boxplot(
        df[columna].dropna(),
        orientation="horizontal"
    )

    plt.title(titulo, fontsize=14)

    plt.grid(alpha=0.25)

    plt.tight_layout()

    plt.show()
```

## 8.1 Distribución de edades

```python
labels_Q1 = {
    1.0: "11 años o menos",
    2.0: "12 años",
    3.0: "13 años",
    4.0: "14 años",
    5.0: "15 años",
    6.0: "16 años o más"
}

labels_Q2 = {
    1.0: "Masculino",
    2.0: "Femenino"
}

labels_Q3 = {
    1.0: "7.º grado",
    2.0: "8.º grado",
    3.0: "9.º grado"
}
```

```python
def grafico_barras_etiquetas(df, columna, titulo, etiquetas=None,
                             xlabel="", ylabel="Frecuencia"):

    datos = df[columna].copy()

    if etiquetas is not None:
        datos = datos.map(etiquetas)

    datos = datos.fillna("Sin dato")

    conteo = datos.value_counts().sort_index()

    plt.figure(figsize=(9,5))

    plt.bar(conteo.index.astype(str), conteo.values)

    plt.title(titulo, fontsize=14)

    plt.xlabel(xlabel)

    plt.ylabel(ylabel)

    plt.grid(axis="y", alpha=0.25)

    plt.tight_layout()

    plt.show()
```

```python
grafico_barras_etiquetas(df, "Q1", "Distribución de edades", labels_Q1, "Edad")
grafico_barras_etiquetas(df, "Q2", "Distribución por sexo", labels_Q2, "Sexo")
grafico_barras_etiquetas(df, "Q3", "Distribución por grado escolar", labels_Q3, "Grado")
```

### Análisis Antropométrico

En esta sección se analizan las variables relacionadas con las características físicas de los estudiantes.

Variables:

- Estatura (Q4)
- Peso (Q5)

Posteriormente se calculará el Índice de Masa Corporal (IMC), el cual será utilizado como variable objetivo del modelo de regresión.

```python
def resumen_estadistico(df, columna):

    serie = df[columna].dropna()

    resumen = pd.DataFrame({
        "Media":[serie.mean()],
        "Mediana":[serie.median()],
        "Desv.Std":[serie.std()],
        "Mínimo":[serie.min()],
        "Máximo":[serie.max()],
        "Q1":[serie.quantile(.25)],
        "Q3":[serie.quantile(.75)]
    })

    return resumen.round(2)
```

### Ingeniería de Características

En esta sección se construye la variable Índice de Masa Corporal (IMC), la cual será utilizada como variable objetivo del modelo de regresión.

El IMC se calcula mediante la fórmula:

IMC = Peso / (Estatura²)

```python
# ======================================================
# Cálculo del Índice de Masa Corporal (IMC)
# ======================================================

df["IMC"] = df["Q5"] / (df["Q4"] ** 2)
```

```python
def clasificar_imc(imc):

    if pd.isna(imc):
        return np.nan

    if imc < 18.5:
        return "Bajo peso"

    elif imc < 25:
        return "Normal"

    elif imc < 30:
        return "Sobrepeso"

    else:
        return "Obesidad"
```

```python
df["Categoria_IMC"] = df["IMC"].apply(clasificar_imc)
```

### Análisis Bivariado

En esta sección se estudian las relaciones entre las principales variables del estudio.

El objetivo es identificar patrones que posteriormente servirán para la construcción de modelos predictivos.

```python
def boxplot_categoria(df, categoria, variable, etiquetas=None,
                      titulo=""):

    datos = df[[categoria, variable]].copy()

    datos = datos.dropna()

    if etiquetas is not None:
        datos[categoria] = datos[categoria].map(etiquetas)

    grupos = [
        datos[datos[categoria] == x][variable]
        for x in datos[categoria].unique()
    ]

    plt.figure(figsize=(8,5))

    plt.boxplot(
        grupos,
        tick_labels=datos[categoria].unique()
    )

    plt.title(titulo)

    plt.ylabel(variable)

    plt.grid(alpha=0.3)

    plt.tight_layout()

    plt.show()
```

```python
boxplot_categoria(
    df,
    "Q2",
    "IMC",
    etiquetas=labels_Q2,
    titulo="Distribución del IMC según sexo"
)
```

```python
boxplot_categoria(
    df,
    "Q3",
    "IMC",
    etiquetas=labels_Q3,
    titulo="Distribución del IMC según grado"
)
```
